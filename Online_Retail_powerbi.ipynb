```python
import pandas as pd
import numpy as np
```

```python
file_path = r"C:\Users\jingl\NSS Advanced PBI\online retails\Online_Retail.csv"
```

```python
try:
  
    df = pd.read_csv(file_path, encoding='ISO-8859-1')
    print("sucessed")
    print(f"dataset: {df.shape[0]} row, {df.shape[1]} column")
except FileNotFoundError:
    print("failed")
```

```python
print("\n--- NULL ---")
print(df.isnull().sum())
```

```python
df_cleaned = df.dropna(subset=['CustomerID'])
```

```python
df_cleaned = df_cleaned[(df_cleaned['Quantity'] > 0) & (df_cleaned['UnitPrice'] > 0)]
```

```python
df_cleaned['TotalAmount'] = df_cleaned['Quantity'] * df_cleaned['UnitPrice']
```

```python
print(f"\nDatacleaned: {len(df_cleaned)} row")
print(f"TotalAmount: ${df_cleaned['TotalAmount'].sum():,.2f}")
```

```python
df_cleaned.head(10).to_csv('cleaned_preview.csv', index=False)
```
